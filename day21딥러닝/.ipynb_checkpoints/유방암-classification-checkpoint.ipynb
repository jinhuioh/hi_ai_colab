```python
# 필수 라이브러리
import pandas as pd
import numpy as np
import random
import tensorflow as tf

# 랜덤 시드 고정
SEED=12
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)  
print("시드 고정: ", SEED)
```

```python
data = pd.read_csv('./유방암df.csv')
data.shape
```

```python
data[:3]
```

```python
data.isnull
```

```python
data['label'].value_counts()
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
#딥러닝 기본에서는 target이 명목형인 경우, dummy로 만들어줘야 한다.(one-hot인코딩) 
```

```python
y_data = to_categorical(data['label'])
y_data[:3]
```

```python
#train시킬 문제
X_data = data.loc[:,:'label']
X_data
```

```python
len(data.columns)
```

```python
X_data.shape
```

```python
# min_max스케일링
# 피처 스케일링
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_data)
X_data_scaled = scaler.fit_transform(X_data)
```

```python
X_data_scaled
```

```python
# train/test data분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data_scaled, y_data, 
                                                    test_size=0.2, 
                                                    shuffle=True, 
                                                    random_state=100)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# 신경망 모델을 만들어보자

```python
# 모델생성
# 심층 신경망 모델
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_dim = 31))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(2, activation='softmax'))#가능성이2 
```

```python
##이 모델로 설정을 하겠다(compile)
model.compile(optimizer='adam', loss='categorical_crossentropy',#회귀에는 loss옵션값 다르게 줘야함!
             metrics=['acc','mae']
             )
```

```python
#설정한 레이어들 요약해서 보기.
model.summary()
```

```python
#batch_size는 몇개 학습하고 업데이트하고 다시 학습할것인지 정하는 것
history = model.fit(X_train, y_train, epochs=200, batch_size=50, verbose=2)
```

```python
test_result = model.evaluate(X_test, y_test)
test_result
```

```python
y_pred = model.predict(X_test)
y_pred[0]
```

# 시각화

```python
import matplotlib.pyplot as plt
plt.plot(history.history['mae'])
plt.plot(history.history['loss'])
plt.plot(history.history['acc'])
plt.legend(['mae','loss','acc'])
plt.grid()
plt.show()
```
