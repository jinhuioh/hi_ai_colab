```python
import pandas as pd
import numpy as np
import random
import tensorflow as tf
print(tf.__version__)
```

```python
#사이킷런에 들어있는 보스턴 집값 데이터를 불러오자.
from sklearn import datasets
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
diabetes = datasets.load_diabetes()
#sklearn.untils.Bunch: dictionary객체로 만들어 놓은 것
type(diabetes)
```

```python
diabetes.keys()
```

```python
#문제
X_data = diabetes.data
X_data
```

```python
#답(당뇨병)
y_data = diabetes.target
y_data
```

```python
diabetes.feature_names
```

```python
X_data.shape
```

```python
y_data.shape
```

```python
#전처리(스캐일링)
scaler = MinMaxScaler()
X_data_scaled = scaler.fit_transform(X_data)
X_data_scaled
```

```python
#y값은 스케일링 하지 않는다.(답이기 때문)
X_train, X_test, y_train, y_test = train_test_split(X_data_scaled, y_data, 
                                                    test_size=0.1, 
                                                    shuffle=True, 
                                                    random_state=11)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Layer를 순서대로 쌓아줄 객체 필요 Sequantial
# 각 Layer는 Dense객체
# 심층 신경망
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#레이어를 순서대로 쌓을 수 있도록 객체를 생성
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=10))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))##회귀이기 때문에 linear
##회귀일때, 마지막 출력층의 활성화함수

##쌓은 것으로 모델 셋팅을 하세요
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
# 모델에 세팅한 내용을 요약해서 프린트
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=500, batch_size=32, verbose=2)
history
```

```python
#평가비교해보자
eval_test = model.evaluate(X_test, y_test)
eval_test
```

```python
pred_test = model.predict(X_test)
pred_test[:5]
```

```python
y_test[:5]
```

```python
#R2는 1에 가까울수록 좋다.
from sklearn.metrics import r2_score
R2 = r2_score(y_test, pred_test, multioutput='variance_weighted')
R2
```
