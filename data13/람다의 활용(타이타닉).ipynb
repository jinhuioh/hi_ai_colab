```python
##apply,map의 용도는 df의 컬럼의 값을 특정한 처리 후, df의 컬럼으로 다시 갱신하거나, 파생변수를 만들어 집어넣기 위한 것.
## => apply, map의 함수로 사용되는 것은 입력값이 시리즈의 값 하나를 처리하는 단위로 만들어야함.
## => return을 반드시 넣어서 함수로 만들어야 함.
```

```python
df = pd.read_csv("../csv_file/titanic_train (1).csv")
```

```python
df.head()
```

```python
# 1.p_class column update=> 1:low,2:middle,3:high
```

```python
def p_class_fun(x):
    if x==1:
        return 'low'
    elif x==2:
        return 'middel'
    else:
        return 'high'
```

```python
df['pclass'] = df['pclass'].apply(p_class_fun)
df.head()
```

```python
# 2.name으로 파생변수 name_result를 만드세요. 글자수를 세어 15이상이면 long, 나머지 short  
```

```python
name_lam = lambda x:'long' if len(x)>=15 else 'short'
```

```python
df['name_result'] = df['name'].apply(name_lam)
df.head()
```

```python
#3. survived 1이면 생존, 0이면 사망, survived column update
```

```python
survived_lam = lambda x:'생존' if x==1 else '사망'
```

```python
df['survived'] = df['survived'].apply(survived_lam)
df.head()
```

```python
#4. age 10대, 20대 , 30대, 40대, 50대, 60대 이상으로 column update
```

```python
#일단 nan값을 평균으로 바꿔보자.
```

```python
df['age'].isnull().sum()
```

```python
df['age'].mean()
```

```python
df['age'] = df['age'].fillna(30)#fillna함수를 이용하여 결측값 없앰
```

```python
df.head(10)
```

```python
def age_fun(x):
    if x<10:
        return '유아'
    elif 10<=x<20:
        return '10대'
    elif 20<=x<30:
        return '20대'
    elif 30<=x<40:
        return '30대'
    elif 40<=x<50:
        return '40대'
    elif 50<=x<60:
        return '50대'
    elif 60<=x<70:
        return '70대'
    else:
        return '그 외'
```

```python
df['age'] = df['age'].apply(age_fun)
df.head()
```

```python
#5. name에 Mr, Dr 를 포함하고 있으면 ‘남자', Miss를 포함하고 있으면 ‘여자', 나머지는 ‘기타' 
```

```python
def name_fun(x):
    if 'Mr' in x:
        return '남자'
    elif 'Dr' in x:
        return '남자'
    elif 'Miss' in x:
        return '여자'
    else:
        return '기타'
```

```python
# #람다로 해보자....
# name_fun1 = (lambda x:'남자' if 'Mr'and 'Dr' in x  else '여자')
```

```python
df['name'] = df['name'].apply(name_fun)
df.head()
```
