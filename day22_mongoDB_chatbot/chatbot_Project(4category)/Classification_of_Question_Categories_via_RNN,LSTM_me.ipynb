# 토큰화(불용어제거)

```python
#한글과 공백을 제외하고 모두 제거
data_c['내용'] = data_c['내용'].str.replace("[^ㄱ-ㅎ ㅏ-ㅣ가-힣]","")
data_c.tail()
```

```python
#전처리 후 테스트용 샘플의 개수
print(len(data_c['내용']))
```

```python
stopwords = ['의','가','이','은','들','는','좀','잘','과','도','를','으로','자',
             '에','와','한','하다','네','구요','데','요']
```

```python
okt = Okt()
```

```python
#불용어제거
data1 = []
for sentence in tqdm(data_c['내용']):
    tokenized_sentence = okt.morphs(sentence, stem=True)#토큰화
    stopwords_removed_sentence = [word for word in tokenized_sentence
                                  if not word in stopwords]#불용어제거
    data1.append(stopwords_removed_sentence)
```

```python
data0 = data1
```

# 단어집합의 크기를 구해보자.
## (집합의 크기로 data 를 fit 시키기 위함)

```python
threshold = 5
total_cnt = len(tokenizer.word_index)
rare_cnt = 0
total_freq = 0
rare_freq = 0

for key, value in tokenizer.word_counts.items():
    total_freq= total_freq + value
    if(value < threshold):
        rare_cnt = rare_cnt + 1
        rare_freq = rare_freq + value
vocab_size= total_cnt-rare_cnt + 1
print("단어의 집합 크기:", vocab_size)
```

```python
#전체 데이터 길이는 7로 맞춘다.
max_len = 10
data1_one_hot = pad_sequences(data1, maxlen = max_len, padding='post')
data1_one_hot[:5]
```

```python
#y데이터 라벨인코딩
# 라벨 인코딩(카테고리형 문자를 숫자로 자동 변환해주는 역할)
from sklearn.preprocessing import LabelEncoder
y = np.array(data_y)
encoder = LabelEncoder()
label = encoder.fit_transform(data_y)
label[:50]
#상품추천:1
#구매목록:0
#의류문의:2
```

```python
encoder.classes_
```

```python
label.astype('float')
```

```python
x = data1_one_hot
y_label = label
X_train, X_test, y_train, y_test = train_test_split(x,
                                                    y_label,
                                                    test_size= 0.2,
                                                    random_state=55
                                                   )
print(X_train.shape, X_test.shape)
```

```python
from tensorflow.keras.layers import Embedding, Dense, LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

# LSTM 으로 분류시작

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=5)
mc = ModelCheckpoint('best_model.h6', monitor='val_acc', mode='max', verbose=1,
                    save_best_only=True)
```

```python
#tokenizer.word_counts.items()
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 16 ,input_length=10))
model.add(LSTM(8))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
             metrics=['acc'])
```

```python
X_train[0]
```

```python
history = model.fit(X_train,
                    y_train,
                    epochs=10,
                    callbacks=[es, mc],
                    batch_size= 80,
                    validation_split=0.2
                    )
```

```python
X_train.shape, y_train.shape
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1) ## 1행2열의 도표, 1열
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model acc.')
plt.xlabel('epoch')
plt.ylabel('acc')
plt.legend(['train_acc','test_acc'])

plt.subplot(1, 2, 2) ## 1행2열의 도표, 2열
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train_loss','test_loss'])
```

```python
def category_predict(new_sentence):
    new_sentence = re.sub(r"[^ㄱ-ㅎ ㅏ-ㅣ가-힣]","", new_sentence)
    new_sentence = okt.morphs(new_sentence, stem=True)#토큰화
    new_sentence = [word for word in new_sentence if not word in stopwords]#불용어제거
    encoded = tokenizer.texts_to_sequences([new_sentence])#정수인코딩
    pad_new = pad_sequences(encoded, maxlen=max_len)#패딩
    score = find_user_category.predict(pad_new)#예측
    #상품추천:1
    #구매목록:0
    #의류문의:2
#     out_x = np.where(score<0.5, "구매목록 내용",(1.5 < score,"상품추천이군요~","의류문의군요~"))

#     print(out_x)
    if(1.5 < score):
        print("상품추천에 대한 내용")
    elif(score < 0.5):
        print("구매목록에 대한 내용")
    elif:
        print("의류문의군요~")
    else:
        pass
```
