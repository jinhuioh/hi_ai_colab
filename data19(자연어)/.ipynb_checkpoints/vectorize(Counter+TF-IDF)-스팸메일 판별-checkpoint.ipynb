```python
from sklearn.feature_extraction.text import *
##CountVectorizer,TfidfVectorizer 사용, 특성추출
#CountVectorizer : 각 텍스트에서 단어 출현 횟수를 카운팅한 벡터
#TfidfVectorizer : TF-IDF라는 값을 사용하여 CountVectorizer의 단점을 보완함
#                   빈도수와 역빈도수를 함께 고려
import warnings
warnings.filterwarnings(action='ignore')
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

### extraction!(reg) => not null, word

```python
#결측치를 빼보자.
spam_df = spam_df[spam_df['v2'].notnull()]
spam_df
```

```python
spam_df.info()
```

```python
#필요없는 컬럼을 지워보자.
spam_df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],
             inplace=True)
```

```python
#중복값 제거
df.drop_duplicates(subset=['v2'], inplace=True)
df
```

```python
spam_df.head()
```

```python
#정규식을 사용해서 영어제외 전부 없애보자.(특수문자 등)


spam_df['v2'] = spam_df['v2'].apply(lambda x: re.sub(r'[^a-zA-Z]+'," ",x))
spam_df.head()
```

```python
#test,train데이터를 나누어보자.
```

```python
spam_df.iloc[:,1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(spam_df.iloc[:,1],#문제 
                                                    spam_df.iloc[:,0],#정답 
                                                    test_size=0.4, 
                                                    random_state=42)
```

```python
print('X_train: ',X_train.shape, 'y_train: ',y_train.shape, 'X_test: ',X_test.shape,'y_test: ',y_test.shape)
```

```python
#fit 해서 정확도 채크해보자.
```

```python
#okt객체 생성
okt = Okt()
```

```python
tfidf = TfidfVectorizer(tokenizer= okt.morphs , 
                        ngram_range=(1, 2), 
                        min_df = 3, 
                        max_df = 0.9)
tfidf
```

```python
#단어 토큰으로 끊어 1,2개로 카운트 한 것을 가지고 공부시키자.
#사전 만들 때 나, 스케일링시 fit()시킬 때, 반드시 train데이터를 사용한다.
#test데이터만을 가지고 fit()시키지 않을 때
tfidf.fit(X_train)
```

```python
#원핫인코딩으로 바꿔보자.(transform)
spam_tfidf_trainx = tfidf.transform(X_train)
spam_tfidf_trainy = tfidf.transform(y_train)
spam_tfidf_trainx
```

```python
# 테스트하기 위해서 테스트 데이터도 1)fit, transform 2) transform(*******)
# 테스트데이터를 tfidf vectorize할때는 train데이터로 fit한 후, transform시켜야 함.
spam_tfidf_testx = tfidf.transform(X_test)
spam_tfidf_testy = tfidf.transform(y_test)
spam_tfidf_test
```
