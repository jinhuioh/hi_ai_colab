```python
con = request.urlopen('https://www.musinsa.com/search/musinsa/goods?q=%ED%8B%B0%EC%85%94%EC%B8%A0&list_kind=small&sortCode=pop&sub_sort=&page=1&display_cnt=0&saleGoods=&includeSoldOut=&setupGoods=&popular=&category1DepthCode=&category2DepthCodes=&category3DepthCodes=&selectedFilters=&category1DepthName=&category2DepthName=&brandIds=&price=&colorCodes=&contentType=&styleTypes=&includeKeywords=&excludeKeywords=&originalYn=N&tags=&campaignId=&serviceType=&eventType=&type=&season=&measure=&openFilterLayout=N&selectedOrderMeasure=&shoeSizeOption=&groupSale=&d_cat_cd=&attribute=')
con 
#http response(header + body)
#header : http result status code
#         (ok: 200, not found: 404, error: 500, permission: 403)
```

```python
con.status #status code 
#con.status를 쓰면내가 요청한 url받아오겠다는 의미.
#내가 요청한 url받는데 성공하면 200이 뜸! 
```

```python
doc = BeautifulSoup(con, 'html.parser')
#HTML Parse란 HTML 문법 규칙에 따른 문자열을, 
#해당 문법을 바탕으로 단어의 의미나 구조를 분석하는 것을 의미
#doc
```

```python
a_tag_list = doc.select('#searchList > li > div.li_inner > div.article_info > p.list_info > a')
a_tag_list
```

```python
a_tag_list[0]
```

```python
a_tag_list[0].text
```

```python
#a태그에서 title만 추출
def tag_source(rlist):
    returnlist=[]
    for i in range(0, len(rlist), 1):
        a= rlist[i].get('title')
        returnlist.append(a)
    return returnlist
```

```python
title_tag_list = tag_source(a_tag_list)
title_tag_list
```

```python
df = pd.DataFrame(data=title_tag_list, columns=['후디'])
df
```

```python
df.to_csv("티셔츠.csv", mode='w', encoding='utf-8')
```
