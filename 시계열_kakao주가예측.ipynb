```python
# 시계열데이터는 시간적인 순서가 중요!
# 미래에 대한 수치를 예측하는 것!
# 최근 5일치를 검증용 데이터로 사용하게 됨.
# 5일치를 제외한 나머지 데이터를 훈련데이터로 사용함.
```

```python
#그래프 그릴때 x축을 시간으로 하기위해 Date컬럼을 디타입이 date으로 되도록 바꿔주자.
kakao_df['Date'] = pd.to_datetime(kakao_df['Date'])
kakao_df.info()
```

```python
#train데이터
kakao_train_df = kakao_df.iloc[:2022]
kakao_train_df[:2022]
```

```python
kakao_train_df.set_index('Date', inplace=True)
kakao_train_df.head(2)
```

```python
kakao_train_df.info()
```

```python
kakao_train_df.plot()
```

```python
## ARIMA(아리마 모델) : 시계열의 유명한 모델
## AR+I+MA
## AR : 값들의 흐름의 특징을 가지고 예측하는 모델
## MA : 값들의 차이의 특징을 가지고 예측하는 모델
## I : AR+MA
## 월별, 계절별 trend도 예측할 수 있도록 해줌.
```

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
```

```python
#아리마 객체생성해보자
model = ARIMA(kakao_train_df['High'].values, order=(2, 0, 10))#order(ar의 값 1의 값 ma의 값)보통 2,1,2을 많이씀 #양이 많은 경우 차이를 더 크게 주자
 #시리즈는 인덱스와 벨류즈의 결합
```

```python
model_fit = model.fit(trend='c', full_output=True)
```

```python
fig = model_fit.plot_predict()#예측한것을 그림으로 그려보자.y:실제값 forecast:예측값
```

```python
forecast_data = model_fit.forecast(steps=5)
forecast_data
```

```python
# 마지막 5일의 예측 데이터
pred_y = forecast_data[0]
pred_y
```

```python
r2_score(test_y, pred_y)
```

```python
from fbprophet import Prophet#이 모델을 쓰려면 names=['ds','y']로 해줘야함.
```

```python
kakao_df2 = pd.read_csv("/content/035720.KS(kakao_finance8years).csv", encoding='utf-8')
kakao_df2
```

```python
kakao_df2 = kakao_df2.drop(columns=['Open','Low','Close','Adj Close','Volume'])
kakao_df2.head()
```

```python
kakao_df2.columns=['ds','y']
kakao_df2
```

```python
kakao_df2['ds'] = pd.to_datetime(kakao_df2['ds'])
kakao_df2.info()
```

```python
prophet = Prophet(
   seasonality_mode='multiplicative',
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=True

)
```

```python
prophet.fit(kakao_df2)
```

```python
furture_data = prophet.make_future_dataframe(periods=5, freq='d')
forecast_data3 = prophet.predict(furture_data)
forecast_data3 
#High 예측한 값
forecast_data3[['ds',	'yhat',	'yhat_lower',	'yhat_upper']].tail(5)#끝에서 5개 예측값만 보기
```

```python
fig3 = prophet.plot(forecast_data3)
```

```python
fig4 = prophet.plot_components(forecast_data3)
```
